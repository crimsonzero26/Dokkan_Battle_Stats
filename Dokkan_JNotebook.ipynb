```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
dokkan_df=pd.read_csv('DokkanStatistics-MasterSheet.csv', header=None)
```

```python
dokkan_df.head()
```

```python
dokkan_df.columns=dokkan_df.iloc[1]
dokkanClean1_df=dokkan_df[:]
dokkanClean1_df.head()
```

```python
dokkanClean2_df=dokkanClean1_df.loc[:, dokkanClean1_df.columns.notna()]
dokkanClean2_df.head()
```

```python
dokkanClean3_df=dokkanClean2_df.drop([0,1])
dokkanClean3_df.head()
```

```python
cols=[4,6,8,10,12,14,16,18,20]

dokkanClean3_df.drop(dokkanClean3_df.columns[cols],axis=1,inplace=True)
dokkanClean3_df.head()
```

```python

dokkanFinal_df=dokkanClean3_df.replace({',':''}, regex=True)
dokkanFinal_df.head(50)
```

```python
dokkanFinal_df.info()
```

```python
dokkanFinal_df['_100HP']=pd.to_numeric(dokkanFinal_df['_100HP'], errors='coerce').convert_dtypes()
```

```python
dokkanFinal_df['_100HP'].dtype
```

```python
dokkanFinal_df['_100HP'].mean()
```

```python
dokkanFinal_df['_100ATK']=pd.to_numeric(dokkanFinal_df['_100ATK'], errors='coerce').convert_dtypes()
dokkanFinal_df['_100DEF']=pd.to_numeric(dokkanFinal_df['_100DEF'], errors='coerce').convert_dtypes()
```

```python
dokkanFinal_df['_55HP']=pd.to_numeric(dokkanFinal_df['_55HP'], errors='coerce').convert_dtypes()
dokkanFinal_df['_55ATK']=pd.to_numeric(dokkanFinal_df['_55ATK'], errors='coerce').convert_dtypes()
dokkanFinal_df['_55DEF']=pd.to_numeric(dokkanFinal_df['_55DEF'], errors='coerce').convert_dtypes()
```

```python
dokkanFinal_df['bHP']=pd.to_numeric(dokkanFinal_df['bHP'], errors='coerce').convert_dtypes()
dokkanFinal_df['bATK']=pd.to_numeric(dokkanFinal_df['bATK'], errors='coerce').convert_dtypes()
dokkanFinal_df['bDEF']=pd.to_numeric(dokkanFinal_df['bDEF'], errors='coerce').convert_dtypes()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=dokkanFinal_df)
plt.xlabel('Dokkan Class')
plt.ylabel('Number of Characters')
plt.title('Characters by Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
c=("#FF0000", "#00FF1A", "#EEFF00", "#2E81FF", "#BB007C")
sns.countplot(x='Type', palette=c, edgecolor='black', data=dokkanFinal_df)
plt.xlabel('Dokkan Types')
plt.ylabel('Number of Characters')
plt.title('Characters by Type')
plt.show()
```

```python
corr_df = dokkanFinal_df[['_100HP', '_100ATK', '_100DEF']]

plt.rcParams['figure.figsize']=(15,9)
heat_map = sns.heatmap(corr_df.corr(), annot = True, linewidths=1, cmap='Blues')
heat_map.set_title(label='Stat Correlation Map', fontsize=25)
heat_map
```
